```python
from tkinter import *

calc = Tk()
num = ""


def insertt(value):
    global num
    num += value
    Resultsbar.config(text=num)


def calculate():
    global num
    try:
        result = eval(num) 
        Resultsbar.config(text=str(result))
        num = str(result) 
    except ZeroDivisionError:
        Resultsbar.config(text="Error (Div by 0)")
        num = ""
    except Exception as e:
        Resultsbar.config(text="Error")
        num = ""


def clear():
    global num
    num = ""
    Resultsbar.config(text="")


Resultsbar = Label(calc, text="")
Resultsbar.grid(row=0, column=0, columnspan=4)


num_1 = Button(calc, text="1", width=5, height=2, command=lambda: insertt("1"))
num_1.grid(row=1, column=0)
num_2 = Button(calc, text="2", width=5, height=2, command=lambda: insertt("2"))
num_2.grid(row=1, column=1)
num_3 = Button(calc, text="3", width=5, height=2, command=lambda: insertt("3"))
num_3.grid(row=1, column=2)
num_4 = Button(calc, text="4", width=5, height=2, command=lambda: insertt("4"))
num_4.grid(row=2, column=0)
num_5 = Button(calc, text="5", width=5, height=2, command=lambda: insertt("5"))
num_5.grid(row=2, column=1)
num_6 = Button(calc, text="6", width=5, height=2, command=lambda: insertt("6"))
num_6.grid(row=2, column=2)
num_7 = Button(calc, text="7", width=5, height=2, command=lambda: insertt("7"))
num_7.grid(row=3, column=0)
num_8 = Button(calc, text="8", width=5, height=2, command=lambda: insertt("8"))
num_8.grid(row=3, column=1)
num_9 = Button(calc, text="9", width=5, height=2, command=lambda: insertt("9"))
num_9.grid(row=3, column=2)
num_0 = Button(calc, text="0", width=5, height=2, command=lambda: insertt("0"))
num_0.grid(row=4, column=0)


plus = Button(calc, text="+", width=5, height=2, command=lambda: insertt("+"))
plus.grid(row=1, column=3)
minus = Button(calc, text="-", width=5, height=2, command=lambda: insertt("-"))
minus.grid(row=2, column=3)
times = Button(calc, text="*", width=5, height=2, command=lambda: insertt("*"))
times.grid(row=3, column=3)
divide = Button(calc, text="/", width=5, height=2, command=lambda: insertt("/"))
divide.grid(row=4, column=3)


equals = Button(calc, text=" = ", width=11, height=2, command=calculate)
equals.grid(row=4, column=1, columnspan=2)

clear_btn = Button(calc, text="C", width=5, height=2, command=clear)
clear_btn.grid(row=4, column=2)

calc.mainloop()
```
